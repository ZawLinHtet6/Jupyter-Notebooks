```python
from PIL import Image, ImageDraw

import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
```

```python
sample_image = Image.open('C:\\Users\\ivenz\\Desktop\\Master\\pedestrian\\train_images\\p2.jpg')
sample_image
```

```python
with open('C:\\Users\\ivenz\\Desktop\\Master\\pedestrian\\train_labels\\p2.xml') as annotate_file:
    print(''.join(annotate_file.readlines()))
```

```python
tree = ET.parse('C:\\Users\\ivenz\\Desktop\\Master\\pedestrian\\train_labels\\p2.xml')
root = tree.getroot()

coors = []
boxes = []

for neighbor in root.iter('bndbox'):
    xmin = int(neighbor.find('xmin').text)
    ymin = int(neighbor.find('ymin').text)
    xmax = int(neighbor.find('xmax').text)
    ymax = int(neighbor.find('ymax').text)
        
    coors.append([xmin, ymin, xmax, ymax])
    
```

```python
label_image = sample_image.copy()

img_bbox = ImageDraw.Draw(label_image)

for bbox in coors:
    img_bbox.rectangle(bbox, outline="cyan", width = 2)
    
label_image
```
