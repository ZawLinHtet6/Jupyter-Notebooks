```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import imageio
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage

# Open annotation file
soup = BeautifulSoup(open('C:\\Users\\ivenz\\Desktop\\Master\\pedestrian\\annotations\\p2.xml'),"lxml")
# import images
image = imageio.imread("C:\\Users\\ivenz\\Desktop\\Master\\pedestrian\\images\\p2.jpg")


# Used to store dimension file bounding box information
bbsOnImg=[]
# Find all nodes that contain the marquee target
for objects in soup.find_all(name="object"):
    # Gets the category name of the current bounding box
    object_name = str(objects.find(name="name").string)
    # Extracting coordinate point information
    xmin = int(objects.xmin.string)
    ymin = int(objects.ymin.string)
    xmax = int(objects.xmax.string)
    ymax = int(objects.ymax.string)
    # Save the information for the bounding box
    bbsOnImg.append(BoundingBox(x1=xmin, x2=xmax, y1=ymin, y2=ymax))
# Initializing the marquee data of imgaug
bbs = BoundingBoxesOnImage( bbsOnImg,shape=image.shape)
# Show results
plt.imshow(bbs.draw_on_image(image, size=3))
```
